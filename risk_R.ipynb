```python
import pandas as pd
import pandas_datareader.data as pdr
import datetime as dt
from datetime import datetime, timedelta
from sklearn import preprocessing
import numpy as np

import seaborn as sns
sns.set()
```

```python
start = dt.datetime(2014,9,17)
end = dt.datetime(2021,5,29)
df = pdr.DataReader('BTC-USD','yahoo',start,end)
```

```python
names = ['High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close', '50SMA', '350SMA', 'R']
```

```python
df['50SMA'] = df.Close.rolling(window=50).mean()
df['350SMA'] = df.Close.rolling(window=350).mean()
df['R'] = df['50SMA']/df['350SMA']
df = df.dropna()

d = preprocessing.normalize(df, axis=0)
normal_df = pd.DataFrame(d, columns=names)
df['R'].plot();
```
